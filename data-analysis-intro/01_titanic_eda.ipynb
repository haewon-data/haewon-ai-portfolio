## üéØ Purpose

To understand survival patterns in the Titanic dataset  
based on variables like sex, passenger class, and age.

## üì¶ Dataset

The dataset used is from [Kaggle](https://www.kaggle.com/c/titanic)  
and publicly available repositories such as Data Science Dojo.

üì¶ 1. Load the Dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid')

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df.head()
```

üìù 2. Column Descriptions
## üìã Feature Overview

| Column | Description |
|--------|-------------|
| PassengerId | Unique ID for each passenger |
| Survived    | Survival status (0 = No, 1 = Yes) |
| Pclass      | Ticket class (1 = 1st, 2 = 2nd, 3 = 3rd) |
| Name        | Full name of the passenger |
| Sex         | Gender |
| Age         | Age in years |
| SibSp       | # of siblings/spouses aboard |
| Parch       | # of parents/children aboard |
| Ticket      | Ticket number |
| Fare        | Passenger fare |
| Cabin       | Cabin number |
| Embarked    | Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton) |

üìä 3. Basic Statistics & Missing Values

```python
# Summary info
df.info()

# Descriptive statistics
df.describe()

# Check missing values
df.isnull().sum()
```

üìà 4. Key Feature Visualizations

```python
# Survival count by sex
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Count by Sex')
plt.show()
```

```python
# Survival count by Pclass
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
# Age distribution
sns.histplot(data=df, x='Age', bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

### üìê Quantifying Insights

To deepen our initial observations,  
we calculate actual survival rates by group.

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title('Survival Rate by Sex')
plt.ylabel('Survival Rate')
plt.show()
```

The survival rate difference between males and females was highly significant.

```python
sns.heatmap(df.pivot_table(index='Pclass', columns='Sex', values='Survived'), annot=True, cmap='Blues')
plt.title('Survival Rate by Pclass and Sex')
plt.show()
```

Even among females, survival rate significantly decreased in 3rd class,
highlighting the impact of socioeconomic status on survival.

üßæ 5. Summary & Insights
## üß† Summary

- **Sex**: Females had a significantly higher survival rate than males, showing a clear gender gap.
- **Pclass**: Survival rates were highest in 1st class.  
  3rd class females showed notably lower survival, reflecting class disparities.
- **Age**: Most passengers were between 20 and 40 years old.

This exploratory analysis helped uncover clear survival patterns across key variables.
